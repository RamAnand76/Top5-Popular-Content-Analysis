```python
import pandas as pd
```

```python
reaction_df = pd.read_csv('Reactions.csv')
reaction_types_df = pd.read_csv('ReactionTypes.csv')
content_df = pd.read_csv('Content.csv')
```

```python
reactions = reaction_df.drop("Unnamed: 0", axis=1)
reactions = reaction_df.drop("Category", axis=1)
reactions
```

```python
reaction_types_df
```

```python
contents = content_df.drop("Unnamed: 0", axis=1)
contents
```

```python
# Merge Content DataFrame with Reaction DataFrame
merged_df = pd.merge(reactions, contents, on='Content ID', how='inner')
merged_df.drop("Unnamed: 0", axis=1)
```

```python
# Merge Reaction Types DataFrame with the merged DataFrame
merged_df = pd.merge(merged_df, reaction_types_df, on="Reaction Type", how='inner', suffixes=('_left', '_right'))
merged_df
```

```python
category_scores = merged_df.groupby('Category')['Score'].sum()
category_scores
```

```python
sorted_category_scores = category_scores.sort_values(ascending=False)
sorted_category_scores
```

```python
top_5_categories = sorted_category_scores.head(5)
top_5_categories
```
